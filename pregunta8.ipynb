```python
def mult_hibrido(n):
    return (n**3 / 3) + (3 / (2 * n**2)) - (5 / 6 * n)
def mult_eliminacion(n):
    return (n**3 / 3) + n**2 - (n/3) 
def mult_gauss(n):
    return (n**3 / 2) + n**2 - (n / 2)

#graficar las funciones
import numpy as np
import matplotlib.pyplot as plt
n = np.linspace(1, 100, 100)
plt.plot(n, mult_hibrido(n), label='Multiplicacin Hbrido')
plt.plot(n, mult_eliminacion(n), label='Multiplicacin Eliminacin')
plt.plot(n, mult_gauss(n), label='Multiplicacin Gauss')
plt.xlabel('Tamao de la matriz (n)')   
plt.ylabel('Nmero de multiplicaciones')
plt.title('Comparacin de mtodos de multiplicacin de matrices')
plt.legend()
plt.grid()
plt.show()
```

```python
def sum_hibrido(n):
    return (n**3/3) + (n**2 / 2) - (5/6*n)
def sum_eliminacion(n):
    return (n**3 / 3) + (n**2 / 2) - ((5*n) / 6) 
def sum_gauss(n): 
    return (n**3 / 2) - (n / 2)
    

#graficar las funciones
n = np.linspace(1, 100, 100)
plt.plot(n, sum_hibrido(n), label='Suma Hbrido')
plt.plot(n, sum_eliminacion(n), label='Suma Eliminacin')
plt.plot(n, sum_gauss(n), label='Suma Gauss')
plt.xlabel('Tamao de la matriz (n)')
plt.ylabel('Nmero de sumas')
plt.title('Comparacin de mtodos de suma de matrices')
plt.legend()
plt.grid()
plt.show()
```
